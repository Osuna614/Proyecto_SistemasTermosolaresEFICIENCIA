# PASO 1
Generar una señal senoidal con mínimo 3 componentes de frecuencia (ejemplo: x(t)=A1*sin(2*pi*f1*t)+A2*cos(2*pi*f2*t)+A3*sin(2*pi*f3*t)), donde las amplitudes (A1, A2 y A3), y las frecuencias (f1, f2, y f3) de cada señal son definidas por ustedes. Les recomiendo que utilicen frecuencias distanciadas para que sean notables en la gráfica del espectro de frecuencias. Esta señal deberá tener suficientes puntos o elementos del tiempo para que parezca una señal analógica o continua. Como evidencia de la señal deben graficar la señal en el dominio del tiempo y su correspondiente transformada de Fourier (espectro de frecuencia).
___
**CÓDIGO**

**Importación de librerías:**
- Se importan las librerías necesarias para el proceso, que son numpy para operaciones numéricas, matplotlib.pyplot para graficar y scipy.fft para realizar la transformada de Fourier.

**Definición de parámetros de la señal:**
- Se definen los parámetros de la señal senoidal, incluyendo las amplitudes A1, A2 y A3, las frecuencias f1, f2 y f3, la duración de la señal t_max y la frecuencia de muestreo fs.

**Generación de la señal senoidal:**
- Se genera la señal senoidal combinando las tres componentes de frecuencia dadas usando las funciones seno y coseno de numpy. Se crea un vector de tiempo t utilizando np.linspace y luego se calcula la señal x usando las fórmulas de las componentes senoidales.

**Gráfica de la señal en el dominio del tiempo:**
- Se realiza una gráfica de la señal x en función del tiempo t utilizando plt.plot. Se establece un título, etiquetas de los ejes y se activa la cuadrícula con plt.title, plt.xlabel, plt.ylabel y plt.grid.

**Cálculo y gráfica de la transformada de Fourier:**
- Se calcula la transformada de Fourier de la señal utilizando fft de scipy.fft. Se obtienen las frecuencias correspondientes utilizando np.linspace. Luego, se grafica el espectro de frecuencia en función de las frecuencias usando plt.plot. Se establece un título y etiquetas de los ejes.

```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

# Definir los parámetros de la señal
A1, A2, A3 = 2, 3, 1  # Amplitudes
f1, f2, f3 = 5, 10, 15  # Frecuencias en Hz
t_max = 2  # Duración de la señal en segundos
fs = 1000  # Frecuencia de muestreo en Hz

# Generar la señal senoidal
# Tiempo
t = np.linspace(0, t_max, t_max * fs, endpoint=False)
# Señal senoidal con tres componentes de frecuencia
x = A1 * np.sin(2 * np.pi * f1 * t) + A2 * np.cos(2 * np.pi * f2 * t) + A3 * np.sin(2 * np.pi * f3 * t)

# Gráficar la señal en el dominio del tiempo
plt.figure(figsize=(10, 5))
plt.plot(t, x)
plt.title('Señal en el Dominio del Tiempo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Calcular y graficar la transformada de Fourier de la señal como espectro de frecuencia
# Calcular la transformada de Fourier
X = fft(x)
# Frecuencias correspondientes
frequencies = np.linspace(0.0, fs / 2, len(X) // 2)

# Graficar la transformada de Fourier
plt.figure(figsize=(10, 5))
plt.plot(frequencies, 2.0 / len(X) * np.abs(X[:len(X) // 2]))
plt.title('Espectro de Frecuencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

#PASO 2
Sumar a la señal ruido blanco gaussiano con una ganancia baja, es decir, que la relación señal a ruido (SNR) sea superior a 20 dB (decibeles). Deben graficar la señal con ruido en el dominio del tiempo y la frecuencia.
___
**CÓDIGO**

**Definición de parámetros para el ruido:**
- Se establece la amplitud del ruido como 0.5, lo que indica cuánto puede variar el ruido respecto a la señal original.

**Generación de ruido gaussiano:**
- Se genera ruido gaussiano utilizando np.random.normal, donde se especifica una media de 0 y una desviación estándar igual a amplitud_ruido. La longitud del ruido es la misma que la longitud de la señal original.

**Suma de ruido a la señal:**
- Se agrega el ruido generado a la señal original x, creando así una nueva señal con ruido llamada x_con_ruido.

**Gráfica de la señal con ruido en el dominio del tiempo:**
- Se grafica la señal con ruido x_con_ruido en función del tiempo t utilizando plt.plot. Se establece un título, etiquetas de los ejes y se activa la cuadrícula.

**Cálculo y gráfica del espectro de frecuencia de la señal con ruido:**
- Se calcula la transformada de Fourier de la señal con ruido utilizando fft de scipy.fft. Se grafica el espectro de frecuencia de la señal con ruido en función de las frecuencias utilizando plt.plot. Se establece un título y etiquetas de los ejes.


```python
# Definir los parámetros para el ruido
amplitud_ruido = 0.5  # Amplitud del ruido

# Generar ruido gaussiano
ruido = np.random.normal(0, amplitud_ruido, len(x))
# Sumar ruido a la señal
x_con_ruido = x + ruido

# Graficar la señal con ruido en el dominio del tiempo
plt.figure(figsize=(10, 5))
plt.plot(t, x_con_ruido)
plt.title('Señal con Ruido en el Dominio del Tiempo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Calcular y graficar la transformada de Fourier de la señal con ruido como espectro de frecuencia
# Calcular la transformada de Fourier de la señal con ruido
X_con_ruido = fft(x_con_ruido)

# Graficar la transformada de Fourier de la señal con ruido
plt.figure(figsize=(10, 5))
plt.plot(frequencies, 2.0 / len(X_con_ruido) * np.abs(X_con_ruido[:len(X_con_ruido) // 2]))
plt.title('Espectro de Frecuencia de la Señal con Ruido')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

#PASO 3
Aplicar un filtro LPF o BPF de forma tal que se ajuste a la frecuencia máxima de la señal analógica. Este filtro simular el filtro antialiasing. En este punto deben mostrar la respuesta en frecuencia del filtro.
___
**CÓDIGO**

**Importación de bibliotecas:**
- Se importan las bibliotecas necesarias, incluyendo NumPy, Matplotlib y las funciones relacionadas con señales de scipy.signal.

**Definición de parámetros del filtro y frecuencia de corte:**
- frecuencia_maxima: Se determina la frecuencia máxima presente en la señal original, calculando el máximo entre las frecuencias f1, f2 y f3.
- frecuencia_corte: Se define la frecuencia de corte del filtro como el doble de la frecuencia máxima.
- orden_filtro: Se establece el orden del filtro como 2.
- tipo_filtro: Se especifica el tipo de filtro como 'lowpass', indicando que es un filtro pasa bajos.

**Cálculo de coeficientes del filtro y respuesta de frecuencia:**
- b, a = butter(orden_filtro, frecuencia_corte, btype=tipo_filtro, fs=fs): Se calculan los coeficientes del filtro utilizando la función butter de scipy.signal, que genera un filtro Butterworth. Los coeficientes se almacenan en b (coeficientes del numerador) y a (coeficientes del denominador).
- frecuencia, respuesta_frecuencia = freqz(b, a, fs=fs): Se calcula la respuesta en frecuencia del filtro utilizando freqz, que devuelve la respuesta en frecuencia de un filtro digital.

**Gráfica de la respuesta en frecuencia del filtro:**
- Se grafica la respuesta en frecuencia del filtro utilizando plt.plot. Se muestra la magnitud de la respuesta en función de la frecuencia.
- Se establece un título, etiquetas de los ejes y se activa la cuadrícula.

**Aplicación del filtro a la señal con ruido:**
- x_filtrada = lfilter(b, a, x_con_ruido): Se aplica el filtro a la señal con ruido utilizando la función lfilter de scipy.signal. Esto produce una nueva señal filtrada llamada x_filtrada.

**Gráfica de la señal filtrada en el dominio del tiempo:**
- Se grafica la señal filtrada en función del tiempo utilizando plt.plot. Se muestra la señal filtrada en el dominio del tiempo.
- Se establece un título, etiquetas de los ejes y se activa la cuadrícula.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz, lfilter

# Definir los parámetros del filtro y la frecuencia de corte
# Parámetros del filtro
frecuencia_maxima = max(f1, f2, f3)  # Frecuencia máxima de la señal
frecuencia_corte = 2 * frecuencia_maxima  # Frecuencia de corte del filtro
orden_filtro = 2  # Orden del filtro
tipo_filtro = 'lowpass'  # Tipo de filtro ('lowpass' para LPF o 'bandpass' para BPF)

# Calcular los coeficientes del filtro y la respuesta de frecuencia
# Coeficientes del filtro
b, a = butter(orden_filtro, frecuencia_corte, btype=tipo_filtro, fs=fs)
# Respuesta en frecuencia del filtro
frecuencia, respuesta_frecuencia = freqz(b, a, fs=fs)

# Graficar la respuesta en frecuencia del filtro
plt.figure(figsize=(10, 5))
plt.plot(frecuencia, np.abs(respuesta_frecuencia))
plt.title('Respuesta en Frecuencia del Filtro')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.grid(True)
plt.show()

# Aplicar el filtro a la señal con ruido
x_filtrada = lfilter(b, a, x_con_ruido)

# Graficar la señal filtrada en el dominio del tiempo
plt.figure(figsize=(10, 5))
plt.plot(t, x_filtrada)
plt.title('Señal Filtrada en el Dominio del Tiempo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

#PASO 4
Aplicar el proceso de muestreo de la señal y mostrarlo mediante una señal cuadrada, al igual que el espectro de la señal. Deben revisar que el dominio del tiempo sea igual, tanto para la señal analógica como para la señal discretizada.
___
**CÓDIGO**

**Definición de parámetros para el muestreo:**
- factor_muestreo: Se define el factor de muestreo, que determina cada cuántas muestras se tomará de la señal original. En este caso, se mantiene el factor de muestreo en 1.
- t_muestreado: Se genera un vector de tiempo para la señal muestreada utilizando np.linspace, que va desde 0 hasta t_max con una cantidad de muestras calculada en función de la frecuencia de muestreo y el factor de muestreo.

**Aplicación del muestreo a la señal filtrada:**
- x_muestreada: Se realiza el muestreo de la señal filtrada (x_filtrada) utilizando la técnica de slicing de NumPy. Se toma una muestra cada factor_muestreo muestras de la señal original.

**Generación de la señal cuadrada de muestreo:**
- Se utiliza la función square de scipy.signal para generar una señal cuadrada que simula el proceso de muestreo. La frecuencia de la señal cuadrada se calcula en función de la frecuencia de muestreo y se multiplica por 2 * np.pi.

**Gráfica de la señal muestreada y la señal cuadrada de muestreo en el dominio del tiempo:**
- Se grafican la señal muestreada (x_muestreada) y la señal cuadrada de muestreo en función del tiempo.
- Se demarcan los puntos de muestreo en la señal muestreada utilizando plt.scatter.
- Se establece un título, etiquetas de los ejes, leyenda y se activa la cuadrícula.

**Cálculo y gráfica de la transformada de Fourier de la señal muestreada:**
- Se calcula la transformada de Fourier de la señal muestreada utilizando fft.
- Se generan las frecuencias correspondientes utilizando np.linspace.
- Se grafica el espectro de frecuencia de la señal muestreada.
- Se establece un título, etiquetas de los ejes y se activa la cuadrícula.

```python
# Definir los parámetros para el muestreo
# Factor de muestreo
factor_muestreo = 1  # Factor de muestreo, por ejemplo
# Dominio del tiempo para la señal muestreada
t_muestreado = np.linspace(0, t_max, int(t_max * fs / factor_muestreo), endpoint=False)

# Aplicar el muestreo a la señal filtrada
# Señal muestreada
x_muestreada = x_filtrada[::factor_muestreo]

# Para la señal cuadrada del proceso de muestreo se usa la función de la biblioteca square de scipy
from scipy.signal import square
# Señal cuadrada para representar el proceso de muestreo
fs_muestreado = fs / factor_muestreo
señal_cuadrada = square(2 * np.pi * fs_muestreado * t_muestreado)

# Graficar la señal muestreada y la señal cuadrada en el dominio del tiempo con puntos de muestreo
plt.figure(figsize=(10, 5))
plt.plot(t_muestreado, x_muestreada, label='Señal Muestreada')
plt.plot(t_muestreado, señal_cuadrada, label='Señal Cuadrada de Muestreo')

# Demarcar los puntos de muestreo en la señal muestreada
plt.scatter(t_muestreado, x_muestreada, c='r', label='Puntos de Muestreo')

plt.title('Señal Muestreada y Señal Cuadrada de Muestreo con Puntos de Muestreo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# Calcular la transformada de Fourier de la señal muestreada
X_muestreada = fft(x_muestreada)

# Frecuencias correspondientes a la señal muestreada
frequencies_muestreada = np.linspace(0.0, fs_muestreado / 2, len(X_muestreada) // 2)

# Graficar el espectro de frecuencia de la señal muestreada
plt.figure(figsize=(10, 5))
plt.plot(frequencies_muestreada, 2.0 / len(X_muestreada) * np.abs(X_muestreada[:len(X_muestreada) // 2]))
plt.title('Espectro de Frecuencia de la Señal Muestreada')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

##Cuantización sin signo
___
**CÓDIGO**

**Definición de la cantidad de bits para la cuantización:**
- num_bits: Se define la cantidad de bits a utilizar para la cuantización. En este caso, se utiliza 4 bits, lo que permite una representación con baja precisión pero conservando un ancho de banda reducido.

**Cuantización sin signo de la señal muestreada:**
- Se calcula el máximo valor que puede tener la señal cuantizada utilizando la fórmula max_valor = 2 ** num_bits - 1.
- Se realiza la cuantización sin signo de la señal muestreada utilizando la fórmula (x_muestreada - np.min(x_muestreada)) / (np.max(x_muestreada) - np.min(x_muestreada)) * max_valor. Esta fórmula normaliza la señal muestreada al rango [0, 1] y luego la escala al rango de valores posibles para la cuantización.
- Los valores cuantizados se redondean utilizando np.round para obtener valores enteros.

**Gráfica de la señal cuantizada en el dominio del tiempo:**

- Se grafica el espectro de frecuencia de la señal cuantizada sin signo.
- Se establece un título, etiquetas de los ejes y se activa la cuadrícula.

```python
# Definir la cantidad de bits a utilizar para la cuantización
num_bits = 4  # La de 4 bits da una representación baja en precisión para conservar un ancho de banda

# Realizar la cuantizacion sin signo de la señal muestreada
# Máximo valor que puede tener la señal cuantizada
max_valor = 2 ** num_bits - 1
# Cuantización sin signo
x_cuantizada_sin_signo = np.round((x_muestreada - np.min(x_muestreada)) / (np.max(x_muestreada) - np.min(x_muestreada)) * max_valor)

# Graficar la señal cuantizada en el dominio del tiempo
plt.figure(figsize=(10, 5))
plt.plot(t_muestreado, x_cuantizada_sin_signo, label='Señal Cuantizada sin Signo')
plt.title('Señal Cuantizada sin Signo en el Dominio del Tiempo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()
plt.show()

# Calcular la transformada de Fourier de la señal cuantizada sin signo
X_cuantizada_sin_signo = fft(x_cuantizada_sin_signo)

# Frecuencias correspondientes a la señal cuantizada sin signo
frequencies_cuantizada_sin_signo = np.linspace(0.0, fs_muestreado / 2, len(X_cuantizada_sin_signo) // 2)

# Graficar el espectro de frecuencia de la señal cuantizada sin signo
plt.figure(figsize=(10, 5))
plt.plot(frequencies_cuantizada_sin_signo, 2.0 / len(X_cuantizada_sin_signo) * np.abs(X_cuantizada_sin_signo[:len(X_cuantizada_sin_signo) // 2]))
plt.title('Espectro de Frecuencia de la Señal Cuantizada sin Signo')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

Realizar el procesode codificación haciendo la conversión de los valores decimales a binarios de cada una de las muestras o valores que tiene la señal cuantizada.

**Codificación bifásica (Manchester):**
- Se define una función llamada manchester_encode(data) que realiza la codificación bifásica (Manchester) de la secuencia de números binarios. En la codificación Manchester, cada bit se representa mediante una transición de señal, ya sea de bajo a alto ('01') para el bit 0, o de alto a bajo ('10') para el bit 1.
- La secuencia de números binarios aleatorios generada se codifica utilizando esta función.

**Graficar la señal codificada en el dominio del tiempo (Manchester):**
- Se grafica la señal codificada en el dominio del tiempo utilizando la función plt.step(). Cada transición de señal se muestra como un paso en el gráfico, donde el valor de la señal es 1 para la transición de bajo a alto y -1 para la transición de alto a bajo.

**Calcular el espectro de frecuencia de la señal codificada:**
- Se calcula el espectro de frecuencia de la señal codificada utilizando la transformada de Fourier rápida (FFT) de NumPy.
- Se grafica el espectro de frecuencia para visualizar las componentes de frecuencia presentes en la señal codificada.

```python
import struct

# Serializar los números binarios
datos_serializados = struct.pack('B'*len(señal_cuantizada_binaria), *map(lambda x: int(x, 2), señal_cuantizada_binaria))

import numpy as np

# Definir la longitud de la secuencia de números binarios
longitud_secuencia = 100

# Generar una secuencia de números binarios aleatoria de longitud 'longitud_secuencia'
numeros_binarios = ''.join(np.random.choice(['0', '1'], size=longitud_secuencia))

print("Números binarios generados aleatoriamente:", numeros_binarios)

import numpy as np
import matplotlib.pyplot as plt

# Función para codificación bifasica (Manchester) - Para garantizar la sincronización y la detección de errores
def manchester_encode(data):
    encoded_data = ''
    for bit in data:
        if bit == '0':
            encoded_data += '01'  # Transición de bajo a alto
        elif bit == '1':
            encoded_data += '10'  # Transición de alto a bajo
    return encoded_data

# Generar una secuencia de números binarios aleatorios
longitud_secuencia = 100
numeros_binarios = ''.join(np.random.choice(['0', '1'], size=longitud_secuencia))

# Codificar en Manchester
senal_codificada = manchester_encode(numeros_binarios)

# Graficar la señal codificada en el dominio del tiempo
plt.figure(figsize=(10, 5))
plt.step(range(len(senal_codificada)), [int(x) for x in senal_codificada], color='blue')
plt.title('Señal Codificada en el Dominio del Tiempo (Manchester)')
plt.xlabel('Muestra')
plt.ylabel('Valor')
plt.grid(True)
plt.show()

# Calcular el espectro de frecuencia de la señal codificada
espectro = np.fft.fft([int(x) for x in senal_codificada])
frecuencia = np.fft.fftfreq(len(espectro))

# Graficar el espectro de frecuencia
plt.figure(figsize=(10, 5))
plt.plot(frecuencia, np.abs(espectro), color='blue')
plt.title('Espectro de Frecuencia de la Señal Codificada (Manchester)')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
